```python
row = 100
column = 100
```

```python
def cwt(signal, wavelet, a0, b0, da):

    ndata = len(signal)

    dt= 1/8000; #fs = 1000
    t = np.arange(0, dt*ndata, dt)
    
    f0=0.849 #frekuensi center
    w0 = 2 * np.pi * f0
    db = (ndata -1)*dt/column
    cwt = np.zeros((row, column))

    a = a0
    for i in range(row):
        b = b0
        for j in range(column):
            real, imaj = wavelet(t, a, b, w0)
            real_cwt = np.sum(np.multiply(signal, real))
            imaj_cwt = np.sum(np.multiply(signal, imaj))
            cwt[i, j] = (np.sqrt(real_cwt**2 + imaj_cwt**2))
            b += db
        a += da
    fc = f0
    
    fk_values = []
    tk_values = []
    
    for i in range(1, row + 1):
        fk = fc / (a0 + (i * da))
        tk = (data_1['x'][ndata - 1] / column) * i
        
        fk = round(fk, 3)  # Membulatkan fk menjadi 3 angka di belakang koma
        tk = round(tk, 3)
        fk_values.append(fk)
        tk_values.append(tk)
        print(f"skala[{i}] = {fk} Hz, waktu[{i}] = {tk} s")
    
    
    return cwt, fk_values, tk_values

```

```python
def morlet_complex(t, a, b, w0):
    real = 1/np.sqrt(a) * 1/(np.pi**0.25) * np.exp(-(t-b)/a*(t-b)/a/2.0) * np.cos(w0*(t-b)/a)
    imaj = 1/np.sqrt(a) * -1/(np.pi**0.25) * np.exp(-(t-b)/a*(t-b)/a/2.0) * np.sin(w0*(t-b)/a)
    return real, imaj
```

```python
def main():
    fig1 = plt.figure(figsize=(15, 6))  # Figur untuk ax1 dan ax3
    fig2 = plt.figure(figsize=(20, 10))  # Figur untuk ax2

    gs1 = gs.GridSpec(3, 1, figure=fig1)  # GridSpec untuk fig1
    gs2 = gs.GridSpec(1, 1, figure=fig2)  # GridSpec untuk fig2

    ax1 = fig1.add_subplot(gs1[0:1, 0])  # Ax1 menggunakan baris pertama (indeks 0) dan kolom pertama (indeks 0)
    ax3 = fig1.add_subplot(gs1[1:, 0], projection="3d")  # Ax3 menggunakan baris kedua hingga ketiga (indeks 1 hingga 2) dan kolom pertama (indeks 0)
    ax2 = fig2.add_subplot(gs2[0, 0])  # Ax2 menggunakan baris pertama (indeks 0) dan kolom pertama (indeks 0)
    
    
    data_1.plot(ax=ax1, x='x', y='y', legend=False, grid=True)
    dt = 1/8000;

    a = float(ent_s.get())#0.0001 #Scallng #(10-4)
    b = (ndata-1)*dt/column
    #b = 4
    da = float(ent_ds.get())#0.0009 #Step Scale #(10-4)

    data1_numpy = data_1['y'].to_numpy()
    result_cwt, fk_values, tk_values = cwt(data1_numpy, morlet_complex, a, b, da)
    # print(np.array(result_cwt).shape)
    
    plot_y = fk_values
    plot_x = tk_values
    xx, yy = np.meshgrid(plot_x, plot_y)
    
    num_yticks = 50
    
    # indeks untuk setiap titik data yang akan ditampilkan labelnya
    y_indexes = np.linspace(0, len(plot_y) - 1, num_yticks, dtype=int)

    # label baru yang sesuai dengan indeks ini
    new_yticklabels = [plot_y[i] for i in y_indexes]

    # buat heatmap dengan label baru ini
    sns.heatmap(result_cwt, cmap='gist_rainbow', ax=ax2, xticklabels=plot_x)
            
    # set xticks and yticks positions to new indexes
    ax2.set_yticks(y_indexes)
    ax2.set_yticklabels(new_yticklabels)

    ax2.invert_yaxis()
    ax2.set_xlabel('Waktu')
    ax2.set_ylabel('Frekuensi')

    ax2.set_title('CWT Contour')
    


    #plt.show()
   
    #ax3.plot_surface(xx, yy, np.array(result_cwt), cmap='gist_rainbow')
    ax3.plot_surface(xx, yy, result_cwt, cmap='gist_rainbow')
    print(f'Jumlah Sample : {ndata}')    
    ymax = max(data_1['y'])
    ymin = min(data_1['y'])
    

    ax1.set_xlim(0, duration)
    ax1.set_ylim(ymin, ymax)
    ax1.set_xlabel('Time(ms)')
    ax1.set_ylabel('Amplitude(mV)')
    ax1.set_title('Input Signal')
    

    ax3.set_xlabel('Waktu')
    ax3.set_ylabel('Frekuensi')
    ax3.set_zlabel('Magnitude (Hz)')
    ax3.set_title('CWT 3D Surface')
    plt.tight_layout()
    plt.show()
```
